# **Unsupervised Classification Model - Yelp data**

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```
reviews = pd.read_csv('yelp_test.csv', header=None)
reviews = reviews[0].tolist()
```

```
reviews
```

```
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(reviews)
```

```
model = KMeans(n_clusters=2, init='k-means++', max_iter=1000, n_init=1)
model.fit(X)
```

```
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```

```
for i in range(2):
  print("Cluster %d:" % i)
  
  for ind in order_centroids[i, :10]:
    print(' %s' % terms[ind])
```

```
print("\n")
print("Prediction")
X = vectorizer.transform(["That movie was terrible. The worst I have ever seen!"])
predicted = model.predict(X)
print(predicted)
```

```
print("\n")
print("Prediction")
X = vectorizer.transform(["That movie was awesome. The best I have ever seen!"])
predicted = model.predict(X)
print(predicted)
```
